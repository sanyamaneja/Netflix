```python
# fill null values of director with Unavailable,country with Unknown,cast with Unknown
# and rating with 'Not given'
net=net.fillna({'Director':'Unavailable','Country':'Unknown','Cast':'Unknown','Rating':'Not given'})
```

```python
# converting into YYMMDD format
net['Date']=pd.to_datetime(net['Release_Date'])
net
```

```python
# creating seperate column for year
net['Year']=net['Date'].dt.year
```

```python
# which year has highest number of release
net['Year'].value_counts()
```

```python
# which year has highest number of release through bar chart
net['Year'].value_counts().plot(kind='bar',edgecolor='black');
```
